```python
def scale_and_assign_metric_skater(scaler,df):
    df2=df.copy()
    for col in df2.columns:
        if df2[col].dtype==float or df[col].dtype==int and col!='season' and col!='age':
            df2[col]=scaler.fit_transform(df2[[col]])
    for i in range(len(df2)):
        df2['mean_mtrc_f'] = df2[['goals', 'assists', 'points','shots', 'hits','faceoff_pct', 
                                             'shot_pct', 'gpg']].mean(axis=1)
        df2['mean_mtrc_d'] = df2[['assists', 'plusmin', 'toi', 'hits', 'block', 
                                             'toi_avg']].mean(axis=1)
    df2=df2[['season','team','name','pos', 'age', 'gp', 'goals', 'assists', 'points', 'plusmin',
       'plm', 'goals_ev', 'goals_pp', 'goals_sh', 'goals_gw', 'assists_ev',
       'assists_pp', 'assists_sh', 'shots', 'toi', 'hits', 'faceoff_wins',
       'faceoff_losses', 'block', 'faceoff_pct', 'shot_pct', 'gpg', 'toi_avg','mean_mtrc_d','mean_mtrc_f']]
    df2=df2.sort_values(by=['season','team','pos'])
    
    mean_dict_frwrd_all={}
    mean_dict_def_all={}
    mean_dict_sktrs_all={}
    
    for s in df2.season.unique():
        mean_dict_frwrd_all[s]=df2[df2.season==s]['mean_mtrc_f'].max()
    
    for s in df2.season.unique():
        mean_dict_def_all[s]=df2[df2.season==s]['mean_mtrc_d'].max()
    
    df2['year_max_d']=df2.season
    
    df2['year_max_f']=df2.season
    
    df2.year_max_f=df2.year_max_f.map(lambda x: mean_dict_frwrd_all[x])

    df2.year_max_d=df2.year_max_d.map(lambda x: mean_dict_def_all[x])

    df2['final_mean_mtrc_d']=round(df2.mean_mtrc_d/df2.year_max_d,2)*100
    
    df2['final_mean_mtrc_f']=round(df2.mean_mtrc_f/df2.year_max_f,2)*100
        
    return df2
```

```python
def scale_and_assign_metric_goalie(scaler,df):
    df2=df.copy()
    
    df2.reset_index(inplace=True)
    
    for col in ['wins','gs','gp','save_pct','shutouts', 'quality_starts','gps',
             'quality_start_pct','gs_above_avg','mpg']:
        df2[col]=scaler.fit_transform(df2[[col]])
        
    for col in ['losses', 'ties', 'goals_avg', 'really_bad_starts', 'ga_pct_minus','plm']:
        df2[col]=scaler.fit_transform((df2[[col]]))
        df2[col]=-1*df2[col]
    
    df2['wins']=df2['wins']*1.1
    df2.gps=df2.gps*0.9
    df2['ties']=0.75*df2['ties']
    df2['save_pct']=2*df2['save_pct']
    df2['gp']=1.5*df2['gp']
    df2['gs_above_avg']=2*df2['gs_above_avg']
    df2['quality_start_pct']=1.1*df2['quality_start_pct']
    df2['really_bad_starts']=2*df2['really_bad_starts']
        
    df2['mean_mtrc']=0
    for i in range(len(df2)):
        df2['mean_mtrc'] = df2[['gp','save_pct','shutouts', 'minutes', 'quality_starts','gps',
             'quality_start_pct','gs_above_avg','mpg','losses', 'ties', 'goals_against','goals_avg', 
                                'really_bad_starts', 'ga_pct_minus','plm']].mean(axis=1)
        
    df2=df2[['season','team','name','age', 'gp', 'gs', 'wins',
       'losses', 'ties', 'goals_against', 'shots', 'saves', 'save_pct',
       'goals_avg', 'shutouts', 'gps', 'minutes', 'quality_starts',
       'quality_start_pct', 'really_bad_starts', 'ga_pct_minus',
       'gs_above_avg', 'goals_scored', 'assists', 'points', 'plm', 'mpg','mean_mtrc']]
    
    df2=df2.sort_values(by=['season','team'])
    
    mean_dict_goalies_all={}
    
    for s in df2.season.unique():
        mean_dict_goalies_all[s]=df2[df2.season==s]['mean_mtrc'].max()
    
    df2['year_mean']=df2.season
    
    df2.year_mean=df2.year_mean.map(lambda x: mean_dict_goalies_all[x])
    
    df2['final_mean_mtrc']=df2.mean_mtrc/df2.year_mean
    
    df2['final_mean_mtrc']=round(df2.final_mean_mtrc*100,2)
    
    return df2
```

```python
def scale_and_assign_metric_nba(scaler,df):
    
    df2=df.copy()
    
    for col in ['gp', 'gs','mp', 'fg', 'fga', 'fg%', '3p', '3pa', '3p%', '2p', '2pa', '2p%',
       'efg%', 'ft', 'fta', 'ft%', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk','tov', 'fouls', 'pts']:
        df2[col]=scaler.fit_transform(df[[col]])
        
    df2['tov']=df2['tov']*-1
    df2['fouls']=df2['fouls']*-0.75    
    df2['3p%']=df2['3p%']*1.15
    df2['efg%']=df2['efg%']*1.1
    
    
    df2['mean_mtrc'] = df2[['gp','mp','3p%','efg%','ft%','trb',
                          'ast','stl','blk','tov','fouls','pts']].mean(axis=1)
    
    mx_mtrc_dict={}
    
    for s in df.Year.unique():
        mx_mtrc_dict[s]=df2[df2.Year==s]['mean_mtrc'].max()
    
    df2['max_mtrc']=df.Year
    
    df2.max_mtrc = df2.max_mtrc.map(lambda x: mx_mtrc_dict[x])
        
    df2=df2.sort_values(by=['Year','team'])
    
    return df2
```
