```python
def find_relative_metrics(year,df_avg,df_rel,year_metric):
    df1=df_avg[df_avg[year_metric]==year]
    df2=df_rel[df_rel[year_metric]==year]
    for col in df2.columns:
        df2[col]=df2[col]/list(df1[col].unique())[0]
    return df2
```

```python
nba_relative_dfs=[]
for i in nba_avg.Year:
    df_to_add = find_relative_metrics(i,nba_avg,nba_mvp_copy,'Year')
    nba_relative_dfs.append(df_to_add)
```
