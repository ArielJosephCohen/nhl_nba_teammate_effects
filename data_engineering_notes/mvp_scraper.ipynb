```python
# imports
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import pandas as pd
from scipy import stats
import numpy as np
import spacy
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from urllib.request import urlopen
import json
import dill
```

```python
url='https://www.hockey-reference.com/awards/hart.html'
```

```python
r=requests.get(url)
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

```python
player_list=[]
for player in soup.find_all('tbody'):
    player_list.append(player.find_all('tr'))
```

```python
player_list=player_list[0]
```

```python
b =[]
for i in range(0,len(player_list)):
    for stat in player_list[i].find_all('td'):
        b.append(stat.text)
```

```python
b[0:18]
```

```python
chunks = [b[x:x+18] for x in range(0, len(b), 18)]
```

```python
chunks[0]
```

```python
season = 2019
for chunk in chunks:
    if season != 2005:
        chunk.append(season)
        season-=1
    else:
        chunk.append(season-1)
        season=season-2
```

```python
chunks[0]
```

```python
nhl_hart = pd.DataFrame(columns=['Player','Team','Season'],index=range(0,len(chunks)))
```

```python
chunks
```

```python
for i in range(len(chunks)):
    nhl_hart.Player[i]=chunks[i][1]
    nhl_hart.Team[i]=chunks[i][3]
    nhl_hart.Season[i]=chunks[i][-1]
```
